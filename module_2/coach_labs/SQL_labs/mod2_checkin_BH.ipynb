### Use SQL to answer the following questions: 

1.  How many rows are in the 'author' database? 

```python
c.execute("""SELECT count(*) as cout_rows FROM books
; """).fetchall() 
```

2. How many authors are called David?  

```python
c.execute("""SELECT COUNT(author) AS DavidSum FROM author WHERE author like 'David%'; """).fetchall() 
```

```python
c.execute("PRAGMA table_info('author')").fetchall()
```

3. How many books each author wrote? 

```python
# Although it was not asked, I ordered by book count in descending order

c.execute("""SELECT author, COUNT(book)
                FROM author 
                GROUP BY author 
                ORDER BY COUNT(book) DESC
        ; """).fetchall() 
```

4. Use both the author table and books table to 
say: how many copies were sold by authors named David 


    a. how many books for each of the Davids
    

```python
c.execute("""SELECT a.author, b.num_copies 
                FROM author a
                JOIN books b
                ON a.book = b.books
                WHERE author LIKE 'David%'
        ; """).fetchall() 
```

    
    b. how many books in total for all Davids

```python
c.execute("""SELECT SUM(b.num_copies)
                FROM author a
                JOIN books b
                ON a.book = b.books
                WHERE author LIKE 'David%'
        ; """).fetchall() 
```
